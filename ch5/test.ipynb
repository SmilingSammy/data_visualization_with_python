```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

path = '/Users/hsy/Library/Fonts/NanumGothic.ttf'
font_name = fm.FontProperties(fname=path, size=10).get_name()
plt.rc('font', family=font_name)
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 부호 깨짐 방지
```

```python
filename = '../data/tips.csv'
tips = pd.read_csv(filename, encoding='utf-8', index_col=0)
```

```python
tips
```

# 1

```python
ax = plt.subplots()
sns.scatterplot(x='total_bill', y='tip', data=tips)
ax = sns.rugplot(x='total_bill', y='tip', data=tips)
ax.set_title('산점도와 Rug Plot')
ax.set_ylabel('팁(tip)')
ax.set_xlabel('총 지급액')
```

# 2

```python
ax = sns.histplot(x='total_bill', data=tips, kde=True)
ax.set_title('총 지급액 히스토그램')
ax.set_ylabel('')
ax.set_xlabel('총 지급액')
```

# 3

```python
tips.head(2)
```

```python
ax = sns.barplot(x='day', y='tip', data=tips, hue='time', errwidth=0)
ax.set_title('요일에 따른 팁(tip)')
ax.set_ylabel('팁(tip)')
ax.set_xlabel('요일')
```

# 4

```python
ax = sns.boxplot(x='day', y='tip', data=tips, hue='time')
ax.set_title('요일에 따른 팁(tip)(boxplot)')
ax.set_ylabel('팁(tip)')
ax.set_xlabel('요일')
```

# 5

```python
ax = sns.violinplot(x='day', y='tip', data=tips, hue='time')
ax.set_title('요일에 따른 팁(tip)(violinplot)')
ax.set_ylabel('팁(tip)')
ax.set_xlabel('요일')
```

# 6

```python
ax = sns.stripplot(x='day', y='tip', data=tips, hue='time', jitter=True)
ax.set_title('요일에 따른 팁(tip)(stripplot)')
ax.set_ylabel('팁(tip)')
ax.set_xlabel('요일')
```

# 7

```python
ax = sns.swarmplot(x='day', y='tip', data=tips, hue='time')
ax.set_title('요일에 따른 팁(tip)(swarmplot)')
ax.set_ylabel('팁(tip)')
ax.set_xlabel('요일')
```
